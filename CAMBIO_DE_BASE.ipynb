# PROGRANA DE CAMBIO DE BASE 10 A 2 - ALGORITMO 1
# List =CamBase10a2_Alg1(arg)

Donde:

*   arg = Parameto de entrada en la funcion, numero natural en base 10
*   List = Lista de coeficienten del numero en base 2 [texto del enlace](https://)

```python
import math
def CamBase10a2_Alg1(num):
  a=[]
  while True:
    r=num%2
    a.append(r)
    q=(num-r)//2
    if(q==0):
      cond=False
      return list(reversed(a))
    num=q
```

# PROGRANA DE CAMBIO DE BASE 10 A 2 - ALGORITMO 2
# List =CamBase10a2_Alg2(arg)

Donde:

*   arg = Parameto de entrada en la funcion, numero natural en base 10
*   List = Lista de coeficienten del numero en base 2

```python
def CamBase10a2_Alg2(num):
  a=[]
  N=num
  while True:
    if(N==0):
      cond=False
      return list(reversed(a))

    if(N%2==0):
      N=(N)/2
      a.append(0)
    else:
      N=(N-1)/2
      a.append(1)
```

```python
CamBase10a2_Alg2(36)
```

#COMPARACION DE RENDIMIENTO DE LOS ALGORITMO 1-2

```python
import time
import matplotlib.pyplot as plt

valor=30
tiempos_algoritmo1=[]
for i in range(1,valor):
  inicio=time.perf_counter()
  CamBase10a2_Alg2(i)
  fin=time.perf_counter()
  tiempos_algoritmo1.append(fin-inicio)

tiempos_algoritmo2=[]
for i in range(1,valor):
  inicio=time.perf_counter()
  CamBase10a2_Alg2(i)
  fin=time.perf_counter()
  tiempos_algoritmo2.append(fin-inicio)

plt.title('Tiempo Transcurrido Algoritmo1-ALgoritmo2',fontsize=13,color="red",style="italic")
plt.plot(tiempos_algoritmo1,label='Algoritmo 1')
plt.plot(tiempos_algoritmo2,label='Algoritmo 2')
plt.legend()
plt.show()
```

# **CONVERSION DE UN NUMERO FRACCIONARIO DE BASE 10 A BASE 2**
# List =CambBaseFrac(arg)

Donde:

*   arg = Parameto de entrada en la funcion, numero decimal tipo '0.abc' en base 10. La parte entera no se trabaja en este programa
*   List = Lista de coeficientes decimales del numero en base 2

```python
def CambBaseFrac(num):
  coef=[]
  k=1
  r=num
  iter=1
  while True:
    if(2*r>=1):
      d=1
    else:
      d=0
    coef.append(d)
    r=2*r-d
    #print(r,iter)
    if(r==0):
      break
    if(iter>15):
      break
    iter=iter+1
  return coef
```
